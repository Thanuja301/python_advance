a poultry farm has group of hens and cows..
test case 1 - 
legs : 16
heads : 6
cows = 2
hens = 4
test case 2 :
legs: 11
heads: 5
cows : no solution
hens: 5

```python
legs = int(input("Enter num of legs: "))
heads = int(input("Enter num of heads: "))
flag = False
for cows in range(0, heads + 1):
    hens = heads - cows
    cal_legs = cows * 4 + hens * 2
    
    if cal_legs == legs:
        flag = True
        break
if flag == True:
    print("Cows :", cows)
    print("Hens :", hens)
else:
    print("No Solution")
```
